```python
# src/fase_5/merge_stock_data.py
import pandas as pd
from pathlib import Path
import re
import numpy as np

RAW_DIR = Path("../data/raw/stock_data/")
OUT_FILE = Path("../data/processed/unitary/stock_data_merged.csv.gz")
OUT_FILE.parent.mkdir(parents=True, exist_ok=True)

def clean_ticker(t):
    if pd.isna(t):
        return None
    t = str(t).strip()
    # eliminar caracteres problemÃ¡ticos y normalizar
    t = re.sub(r'[\^\s/\\]+', '', t)
    t = re.sub(r'[^0-9A-Za-z_.-]', '', t)
    return t

def detect_header_row(df):
    """Detecta la fila que contiene el nombre de la columna 'Date' (o similar)."""
    for i in range(min(6, len(df))):  # buscar en las primeras filas
        row = df.iloc[i].astype(str).str.lower()
        if any(row.str.contains(r'\bdate\b')) or any(row.str.contains(r'\byear\b')):
            return i
    return None

def detect_ticker(df):
    """Intenta detectar el ticker buscando cadenas evidentes en las primeras 3 filas."""
    for i in range(min(3, len(df))):
        for val in df.iloc[i].tolist():
            if pd.isna(val):
                continue
            s = str(val).strip()
            # heurÃ­stica: contiene punto (MC), empieza por ^, todo mayÃºsculas o contiene letras y nÃºmeros
            if re.search(r'\.MC$|^\^|^[A-Z0-9]{1,6}$', s) or re.search(r'[A-Za-z]{2,}\.[A-Za-z]{1,4}', s):
                return s
    # fallback: devolver la primera no-nan del dataframe
    for i in range(min(3, len(df))):
        for val in df.iloc[i].tolist():
            if pd.notna(val) and str(val).strip() and len(str(val).strip()) <= 12:
                return str(val).strip()
    return None

def normalize_df(df, ticker_raw):
    # detectar fila header (donde estÃ¡ 'Date')
    header_row = detect_header_row(df)
    if header_row is None:
        # si no detecta, asumimos que la fila 1 (Ã­ndice 1) es header (segÃºn tu ejemplo)
        header_row = 1 if len(df) > 1 else 0

    # extraer columnas y datos
    new_cols = df.iloc[header_row].fillna("").astype(str).tolist()
    data = df.iloc[header_row+1 : ].copy()
    data.columns = new_cols

    # Si la columna que contiene fecha no se llama exactamente 'Date', buscarla
    date_col = None
    for col in data.columns:
        if isinstance(col, str) and col.strip().lower() in ("date", "fecha", "price"):
            date_col = col
            break
    if date_col is None:
        # si no hay, buscar la columna cuyo contenido parece fecha en la primera filas
        for col in data.columns:
            sample = data[col].dropna().astype(str).head(5).tolist()
            if any(re.match(r'^\d{4}-\d{2}-\d{2}$', s) for s in sample):
                date_col = col
                break

    if date_col is None:
        raise ValueError("No se pudo localizar la columna Date en el archivo (cabecera detectada: {}).".format(header_row))

    # renombrar date_col a 'Date'
    data = data.rename(columns={date_col: "Date"})

    # limpiar ticker y aÃ±adir sufijo a las columnas (excepto Date)
    ticker_clean = clean_ticker(ticker_raw) or "TICKER"
    # si alguna columna vacÃ­a -> rellenar con nombres default
    new_columns = []
    for c in data.columns:
        if not isinstance(c, str) or c.strip() == "":
            # intentar inferir por posiciÃ³n: Close, High, Low, Open, Volume
            new_columns.append(None)
        else:
            new_columns.append(c.strip())
    # si hay nombrado None, asignar por posiciÃ³n
    col_map = {}
    standard = ["Date", "Close", "High", "Low", "Open", "Volume", "Adj Close"]
    for idx, col in enumerate(new_columns):
        if col is None or col == "":
            # asignar por Ã­ndice si posible (asumimos estructura tÃ­pica)
            if idx < len(standard):
                col_map[new_columns[idx]] = standard[idx]
                new_columns[idx] = standard[idx]
            else:
                new_columns[idx] = f"col_{idx}"
        # else keep
    data.columns = new_columns

    # ahora aÃ±adir sufijo a todas las columnas excepto Date
    rename_dict = {}
    for c in data.columns:
        if c == "Date":
            continue
        # limpiar nombre de columna
        c_clean = re.sub(r'\s+', '_', c)
        rename_dict[c] = f"{ticker_clean}_{c_clean}"

    data = data.rename(columns=rename_dict)

    # convertir Date y valores numÃ©ricos
    data["Date"] = pd.to_datetime(data["Date"], errors="coerce")
    # convertir otras columnas a numÃ©rico (suppress errors -> NaN)
    for col in data.columns:
        if col == "Date":
            continue
        data[col] = pd.to_numeric(data[col], errors="coerce")

    # eliminar filas sin Date
    data = data.dropna(subset=["Date"]).reset_index(drop=True)

    # ordenar por Date
    data = data.sort_values("Date").reset_index(drop=True)

    return data

def process_all(raw_dir):
    frames = []
    for fp in sorted(raw_dir.glob("*.csv.gz")):
        print("Procesando:", fp.name)
        try:
            df_raw = pd.read_csv(fp, compression="gzip", header=None, dtype=str)
            ticker_raw = detect_ticker(df_raw)
            df_norm = normalize_df(df_raw, ticker_raw)
            frames.append(df_norm)
        except Exception as e:
            print(f"âš ï¸ Error procesando {fp.name}: {e}")

    # Merge outer por Date
    if not frames:
        raise RuntimeError("No se han procesado archivos correctamente.")
    merged = frames[0]
    for f in frames[1:]:
        merged = pd.merge(merged, f, on="Date", how="outer")

    merged = merged.sort_values("Date").reset_index(drop=True)
    return merged

if __name__ == "__main__":
    merged_df = process_all(RAW_DIR)
    merged_df.to_csv(OUT_FILE, index=False, compression="gzip")
    print("âœ… Merge completado:", OUT_FILE)
    print("Filas:", len(merged_df), "Columnas:", len(merged_df.columns))
    # print merged df
    display(merged_df)
```

```python
import pandas as pd
from pathlib import Path
import re

# Rutas
RAW_DIR = Path(r"..\data\raw\macro_data")
OUT_FILE = Path(r"..\data\processed\unitary\macro_data_merged.csv.gz")
OUT_FILE.parent.mkdir(parents=True, exist_ok=True)

def clean_ticker(name):
    """Limpia y normaliza el nombre base del archivo para usarlo como prefijo."""
    name = re.sub(r'\.csv.*$', '', name)
    name = re.sub(r'[^A-Za-z0-9]+', '_', name)
    return name.strip("_").upper()

def detect_date_column(df):
    """Detecta la columna que contiene fechas."""
    for col in df.columns:
        if re.search(r'date|fecha|time|period', str(col), re.IGNORECASE):
            return col
    # si no hay coincidencias, intenta detectar por contenido
    for col in df.columns:
        sample = df[col].dropna().astype(str).head(5)
        if any(re.match(r'\d{4}-\d{2}-\d{2}', s) for s in sample):
            return col
    return None

def normalize_df(df, name):
    """Normaliza un dataframe: limpia cabeceras, fechas y nombres."""
    # limpiar cabecera
    df.columns = df.columns.astype(str).str.strip()
    df = df.dropna(how="all")

    date_col = detect_date_column(df)
    if not date_col:
        raise ValueError(f"No se encontrÃ³ columna de fecha en {name}")

    df = df.rename(columns={date_col: "Date"})
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    df = df.dropna(subset=["Date"])

    # convertir otras columnas a numÃ©rico si posible
    for col in df.columns:
        if col != "Date":
            df[col] = pd.to_numeric(df[col])

    # renombrar columnas con prefijo
    rename_map = {
        c: f"{name}_{re.sub(r'\\s+', '_', c)}"
        for c in df.columns if c != "Date"
    }
    df = df.rename(columns=rename_map)

    # eliminar duplicadas y ordenar
    df = df.drop_duplicates(subset=["Date"]).sort_values("Date").reset_index(drop=True)
    return df

def process_all(raw_dir):
    frames = []
    for fp in sorted(raw_dir.glob("*.csv.gz")):
        print(f"Procesando: {fp.name}")
        df = pd.read_csv(fp, compression="gzip", header=0, dtype=str)
        ticker = clean_ticker(fp.stem)
        df_clean = normalize_df(df, ticker)
        frames.append(df_clean)

    if not frames:
        raise RuntimeError("No se han podido procesar archivos correctamente.")

    print("ðŸ§© Combinando todos los datasets...")
    merged = frames[0]
    for f in frames[1:]:
        merged = pd.merge(merged, f, on="Date", how="outer")

    merged = merged.sort_values("Date").reset_index(drop=True)
    return merged

if __name__ == "__main__":
    merged_df = process_all(RAW_DIR)
    merged_df.to_csv(OUT_FILE, index=False, compression="gzip")
    print("âœ… Merge completado y guardado en:")
    print(OUT_FILE)
    print("ðŸ“… Filas:", len(merged_df), " | ðŸ“Š Columnas:", len(merged_df.columns))
    display(merged_df)
```

```python
# src/fase_5/commodity_data.py
import pandas as pd
from pathlib import Path
import re
import numpy as np

RAW_DIR = Path("../data/raw/market_risk/")
OUT_FILE = Path("../data/processed/unitary/market_risk_merged.csv.gz")
OUT_FILE.parent.mkdir(parents=True, exist_ok=True)

def clean_ticker(t):
    if pd.isna(t):
        return None
    t = str(t).strip()
    # eliminar caracteres problemÃ¡ticos y normalizar
    t = re.sub(r'[\^\s/\\]+', '', t)
    t = re.sub(r'[^0-9A-Za-z_.-]', '', t)
    return t

def detect_header_row(df):
    """Detecta la fila que contiene el nombre de la columna 'Date' (o similar)."""
    for i in range(min(6, len(df))):  # buscar en las primeras filas
        row = df.iloc[i].astype(str).str.lower()
        if any(row.str.contains(r'\bdate\b')) or any(row.str.contains(r'\byear\b')):
            return i
    return None

def detect_ticker(df):
    """Intenta detectar el ticker buscando cadenas evidentes en las primeras 3 filas."""
    for i in range(min(3, len(df))):
        for val in df.iloc[i].tolist():
            if pd.isna(val):
                continue
            s = str(val).strip()
            if re.search(r'\.MC$|^\^|^[A-Z0-9]{1,6}$', s) or re.search(r'[A-Za-z]{2,}\.[A-Za-z]{1,4}', s):
                return s
    for i in range(min(3, len(df))):
        for val in df.iloc[i].tolist():
            if pd.notna(val) and str(val).strip() and len(str(val).strip()) <= 12:
                return str(val).strip()
    return None

def normalize_df(df, prefix):
    # detectar fila header (donde estÃ¡ 'Date')
    header_row = detect_header_row(df)
    if header_row is None:
        header_row = 1 if len(df) > 1 else 0

    # extraer columnas y datos
    new_cols = df.iloc[header_row].fillna("").astype(str).tolist()
    data = df.iloc[header_row + 1:].copy()
    data.columns = new_cols

    # detectar columna de fecha
    date_col = None
    for col in data.columns:
        if isinstance(col, str) and col.strip().lower() in ("date", "fecha", "price"):
            date_col = col
            break
    if date_col is None:
        for col in data.columns:
            sample = data[col].dropna().astype(str).head(5).tolist()
            if any(re.match(r'^\d{4}-\d{2}-\d{2}$', s) for s in sample):
                date_col = col
                break
    if date_col is None:
        raise ValueError(f"No se pudo localizar la columna Date (cabecera detectada: {header_row}).")

    data = data.rename(columns={date_col: "Date"})

    # rellenar columnas vacÃ­as si las hay
    new_columns = []
    for c in data.columns:
        if not isinstance(c, str) or c.strip() == "":
            new_columns.append(None)
        else:
            new_columns.append(c.strip())
    standard = ["Date", "Close", "High", "Low", "Open", "Volume", "Adj Close"]
    for idx, col in enumerate(new_columns):
        if col is None or col == "":
            if idx < len(standard):
                new_columns[idx] = standard[idx]
            else:
                new_columns[idx] = f"col_{idx}"
    data.columns = new_columns

    # aÃ±adir prefijo del archivo a todas las columnas excepto Date
    rename_dict = {}
    for c in data.columns:
        if c == "Date":
            continue
        c_clean = re.sub(r'\s+', '_', c)
        rename_dict[c] = f"{prefix}_{c_clean}"
    data = data.rename(columns=rename_dict)

    # convertir Date y valores numÃ©ricos
    data["Date"] = pd.to_datetime(data["Date"], errors="coerce")
    for col in data.columns:
        if col != "Date":
            data[col] = pd.to_numeric(data[col], errors="coerce")

    # limpiar y ordenar
    data = data.dropna(subset=["Date"]).reset_index(drop=True)
    data = data.sort_values("Date").reset_index(drop=True)

    return data

def process_all(raw_dir):
    frames = []
    for fp in sorted(raw_dir.glob("*.csv.gz")):
        prefix = fp.stem  # nombre del archivo sin extensiÃ³n
        print(f"Procesando: {fp.name}  â†’  prefijo: {prefix}")
        try:
            df_raw = pd.read_csv(fp, compression="gzip", header=None, dtype=str)
            df_norm = normalize_df(df_raw, prefix)
            frames.append(df_norm)
        except Exception as e:
            print(f"âš ï¸ Error procesando {fp.name}: {e}")

    if not frames:
        raise RuntimeError("No se han procesado archivos correctamente.")

    merged = frames[0]
    for f in frames[1:]:
        merged = pd.merge(merged, f, on="Date", how="outer", suffixes=("", "_dup"))
        merged = merged.loc[:, ~merged.columns.duplicated()]

    merged = merged.sort_values("Date").reset_index(drop=True)
    return merged

if __name__ == "__main__":
    merged_df = process_all(RAW_DIR)
    merged_df.to_csv(OUT_FILE, index=False, compression="gzip")
    print("âœ… Merge completado:", OUT_FILE)
    print("Filas:", len(merged_df), "Columnas:", len(merged_df.columns))
    display(merged_df)
```
